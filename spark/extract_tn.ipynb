```python
import requests
```

```python
districts = requests.get('https://electoralsearch.in/Home/GetDistList?st_code=S22').json()
```

```python
scrape_targets = {}
```

```python
for district in districts:
    scrape_targets[district['dist_no']] = []
```

```python
for district in districts:
    ac = requests.get('https://electoralsearch.in/Home/GetAcList?st_code=S22&dist_no={}'.format(district['dist_no'])).json()
    for ac in ac:
        scrape_targets[district['dist_no']].append(ac['ac_no'])
```

```python
scrape_list = []
for district in scrape_targets:
    for ac in scrape_targets[district]:
        scrape_list.append((int(district), int(ac)))
```

```python
 # Scrape using threadpool executor
from concurrent import futures
from concurrent.futures import ThreadPoolExecutor
import re

def scrape(district, ac):
    print('Scraping', district, ac)
    try:
        parts_list_html = requests.get(f'https://www.elections.tn.gov.in/SSR2022_MR_05012022/ac{ac}.html').text
        with open('out.html', 'w') as f:
            f.write(parts_list_html)
        parts_list = re.findall(rf'https://www.elections.tn.gov.in/rollpdf/Captcha2.aspx\?dt=dt{district}&amp;ac=ac{ac:03d}&amp;pt=ac{ac:03d}([0-9]+)', parts_list_html)
        result = []
        for part in parts_list:
            result.append((district, ac, int(part)))
        return result
    except:
        return None

parts = []
with ThreadPoolExecutor(max_workers=10) as executor:
    for result in executor.map(lambda x: scrape(*x), scrape_list[:10]):
        if result is not None:
            print(result)
```
